# Queries

## Output the number of movies in each category, sorted in descending order. 

```python
from pyspark.sql.functions import count, first
print(df_film.describe())
print(df_film_category.describe())

df_film.join(other=df_film_category, on="film_id").join(other=df_category, on="category_id").groupBy(df_film_category.category_id, df_category.name).agg(count("film_id").alias("movie_count")).select("name", "movie_count").withColumnRenamed("name", "category_name").show()
```

## Output the 10 actors whose movies rented the most, sorted in descending order. 

```python
from pyspark.sql.functions import count, col
print(df_actors.describe())
print(df_film.describe())
print(df_film_actor.describe())
print(df_rental.describe())
print(df_inventory.describe())

df_actors.join(other=df_film_actor, on="actor_id").join(other=df_film, on="film_id").join(other=df_inventory, on="film_id").join(other=df_rental, on="inventory_id").groupby("actor_id", df_actors.first_name, df_actors.last_name).agg(count("rental_id").alias("rentals")).orderBy(col("rentals").desc()).limit(10).select(["first_name", "last_name", "rentals"]).show()
```

## Output the category of movies on which the most money was spent. 

```python
from pyspark.sql.functions import sum
print(df_category.describe())
print(df_film_category.describe())
print(df_rental.describe())
print(df_inventory.describe())
print(df_payment.describe())


df_category.join(df_film_category, on="category_id").join(df_inventory, on="film_id").join(df_rental, on="inventory_id").join(df_payment, on="rental_id").groupBy(df_category.category_id, df_category.name).agg(sum(df_payment.amount).alias("spent_money")).orderBy(col("spent_money").desc()).select(col("name").alias("category_name"), "spent_money").show(1, False)
```

## Output the names of movies that are not in the inventory. 

```python

print(df_film.describe())
print(df_inventory.describe())

df_film.join(df_inventory, "film_id", "left").where(df_inventory.film_id.isNull()).select(df_film.title.alias("non_inventory_movies")).show(1000, False)
```

## Output the top 3 actors who have appeared most in movies in the “Children” category. If several actors have the same number of movies, output all of them. 

```python
from pyspark.sql.functions import count, col, dense_rank
from pyspark.sql.window import Window
print(df_actors.describe())
print(df_film.describe())
print(df_film_actor.describe())
print(df_category.describe())
print(df_film_category.describe())


agg = (
    df_actors
    .join(df_film_actor, on="actor_id")
    .join(df_film, on="film_id")
    .join(df_film_category, on="film_id")
    .join(df_category, on="category_id")
    .where(df_category.name == "Children")
    .groupBy(df_actors.actor_id, df_actors.first_name, df_actors.last_name)
    .agg(count(df_actors.actor_id).alias("appearances"))
)

w = Window.orderBy(col("appearances").desc())

(
    agg
    .withColumn("rnk", dense_rank().over(w))
    .where(col("rnk") <= 3)
    .orderBy(col("appearances").desc())
    .select(df_actors.actor_id, df_actors.first_name, df_actors.last_name, "appearances").show(50)
    
)
```

## Output cities with the number of active and inactive customers (active - customer.active = 1). Sort by the number of inactive customers in descending order. 

```python
from pyspark.sql.functions import count, col, sum, when
print(df_customer.describe())
print(df_city.describe())
print(df_address.describe())



df_customer.join(df_address, on="address_id").join(df_city, on="city_id").groupby(df_city.city_id, df_city.city).agg(count(when(df_customer.active == 1, 1)).alias("active_customers"), count(when(df_customer.active != 1, 1)).alias("inactive_customers")).orderBy(col("inactive_customers").desc()).show(500)




```

## Output the category of movies that have the highest number of total rental hours in the cities (customer.address_id in this city), and that start with the letter “a”. Do the same for cities with a “-” symbol.

```python
import pyspark.sql.functions as F
print(df_category.describe())
print(df_film_category.describe())
print(df_rental.describe())
print(df_inventory.describe())
print(df_payment.describe())
print(df_customer.describe())
print(df_city.describe())
print(df_address.describe())



start_a = (
    df_category
        .join(df_film_category, on="category_id")
        .join(df_film, on="film_id")
        .join(df_inventory, on="film_id")
        .join(df_rental, on="inventory_id")
        .join(df_customer, on="customer_id")
        .join(df_address, on="address_id")
        .join(df_city, on="city_id")
        .where(df_film.title.ilike("a%"))
        .groupBy(df_category.category_id, df_category.name)
        .agg(
            F.sum(
                (F.unix_timestamp(df_rental.return_date) - F.unix_timestamp(df_rental.rental_date)) / 3600
            ).alias("total_rental_hours")
        )
        .orderBy(F.col("total_rental_hours").desc())
        .limit(1)
        .select(F.col("name").alias("category"), "total_rental_hours")
    )

has_hyph =  (
    df_category
        .join(df_film_category, on="category_id")
        .join(df_film, on="film_id")
        .join(df_inventory, on="film_id")
        .join(df_rental, on="inventory_id")
        .join(df_customer, on="customer_id")
        .join(df_address, on="address_id")
        .join(df_city, on="city_id")
        .where(df_city.city.like("%-%"))
        .groupBy(df_category.category_id, df_category.name)
        .agg(
            F.sum(
                (F.unix_timestamp(df_rental.return_date) - F.unix_timestamp(df_rental.rental_date)) / 3600
            ).alias("total_rental_hours")
        )
        .orderBy(F.col("total_rental_hours").desc())
        .limit(1)
        .select(F.col("name").alias("category"), "total_rental_hours")
    )

start_a.show()
has_hyph.show()

```
