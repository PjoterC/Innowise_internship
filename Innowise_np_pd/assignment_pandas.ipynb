## Perform data analysis using Pandas by completing the given tasks.

```python
import pandas as pd
```

```python
# loading the dataset
data = pd.read_csv("./data/adult.data.csv")
data.head()
```

**1. Calculate how many men and women (attribute *sex*) are represented in this dataset.**

```python
counts = data["sex"].value_counts()
print(f"Men: {counts.Male}, Women: {counts.Female}")

```

**2. What is the average age of men (attribute *age*) across the entire dataset?**

```python
men = data.loc[data["sex"] == "Male"]
avg_age = men["age"].mean()
print(avg_age)
```

**3. What is the proportion of United States citizens (*native-country* attribute)?**

```python
us_mask = data["native-country"] == "United-States"
us_proportion = us_mask.mean()

print(f"Proportion of US citizens: {us_proportion*100}%")


```

**4-5. Calculate the mean and standard deviation of the age of those who earn more than 50K per year (the *salary* attribute) and those who earn less than 50K per year**

```python
#print(data.salary.unique())
filter_more = data.loc[data["salary"] == ">50K"]
filter_less = data.loc[data["salary"] == "<=50K"]


more_age_mean = filter_more["age"].mean()
more_age_std = filter_more["age"].std()
less_age_mean = filter_less["age"].mean()
less_age_std = filter_less["age"].std()

print(f">50k age mean: {more_age_mean} years")
print(f">50k age std: {more_age_std} years")
print(f"<=50k age mean: {less_age_mean} years")
print(f"<=50k age std: {less_age_std} years")

```

**6. Is it true that people who earn more than 50k have at least a higher education? (*education indicator – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters or Doctorate*)**

```python
filter_more = data.loc[data["salary"] == ">50K"]
TITLES = ["Bachelors", "Prof-school", "Assoc-acdm", "Assoc-voc", "Masters", "Doctorate"]

elitism_check = filter_more["education"].isin(TITLES).all()
examples = filter_more.loc[~filter_more["education"].isin(TITLES)]
    

print(elitism_check)
#print("Examples:")
#display(examples)
```

**7. Find age statistics for each race (*race* attribute) and each gender. Use *groupby* and *describe*. Find the maximum age of males of the *Asian-Pac-Islander* race.**

```python
grouping = data.groupby(["race", "sex"])
display(grouping.describe())
max_age = grouping["age"].max().loc["Asian-Pac-Islander", "Male"]

print(f"Max age found: {max_age}")

```

**8. Among whom is the proportion of high-earning individuals (>50K) higher: married or single men (*marital-status* indicator)? We consider those whose *marital-status* begins with *Married* (Married-civ-spouse, Married-spouse-absent, or Married-AF-spouse) to be married; the rest are considered single.**

```python
men = data[data["sex"] == "Male"]
is_married = men["marital-status"].str.startswith("Married")

married_rate = (men[is_married]["salary"] == ">50K").mean()
single_rate  = (men[~is_married]["salary"] == ">50K").mean()

print(f"Married men >50K: {married_rate}")
print(f"Single  men >50K: {single_rate}")


```

**9. What is the maximum number of hours a person works per week (hours-per-week indicator)? How many people work that many hours, and what percentage of them earn a lot?**

```python
max_hours = data["hours-per-week"].max()
print(f"Maximum work time: {max_hours}")
people_with_max_hours = data[data["hours-per-week"] == max_hours]
print(f"Amount of people with maximum work time: {len(people_with_max_hours)}")
earn_a_lot = people_with_max_hours[people_with_max_hours["salary"] == ">50K"]
earn_a_lot_ratio = len(earn_a_lot) * 100 / len(people_with_max_hours)
print(f"Of those {earn_a_lot_ratio}% earn a lot of money")
```

**10. Calculate the average working hours (*hours-per-week*) of low and high earners (*salary*) for each country (*native-country*).**

```python
avg_hours = data.groupby(["native-country", "salary"])["hours-per-week"].mean().unstack("salary") # unstack for better display

display(avg_hours)


```

**11.Group people into age groups *young*, *adult*, *retiree*, where:**
* *young* corresponds to 16-35 years old
* *adult* - 35-70 years old
* *retiree* - 70-100 years old

**Enter the name of the appropriate age group for each person in the new AgeGroup column.**

```python
bins = [16, 35, 70, 100]
labels = ['young', 'adult', 'retiree']
data['AgeGroup'] = pd.cut(data['age'], bins=bins, labels=labels)


#display(data[['age', 'AgeGroup']])
display(data.head())
#display(data['AgeGroup'].value_counts())
```

**12-13. Determine the number of people earning >50K in each age group (AgeGroup column), and also output the name of the age group in which people most often earn more than 50K (>50K)**

```python
bins = [16, 35, 70, 100]
labels = ['young', 'adult', 'retiree']
data['AgeGroup'] = pd.cut(data['age'], bins=bins, labels=labels)

# 12. number of people earning >50K in each age group
high_earners = data[data['salary'] == '>50K']
counts = high_earners['AgeGroup'].value_counts().sort_index()
display(counts)

# 13. age group in which people most often earn >50K
top_group = counts.idxmax()
print(f"Age group with the most >50K earners: {top_group}")
```

**14. Group people by employment type (the occupation column) and determine the number of people in each group. Then write a filter function, filter_func, that will return only those groups in which the average age (the age column) is 40 or less and in which all employees work more than 5 hours per week (the hours-per-week column).**

```python
grouping = data.groupby("occupation")

display(grouping["occupation"].describe())


filter_func = lambda x: (x["age"].mean() <= 40) and (x["hours-per-week"] > 5).all()

display(grouping.filter(filter_func))
```
