**1.** Given a random array, change the sign of elements whose values ​​are between 3 and 8

```python
array = np.array(np.random.randint(0, 12+1, size=(1, 10)))
print(array)

# works for any array
nArray = np.where((array > 3) & (array < 8), np.negative(array), array)
print(nArray)
```

**2.** Replace the maximum element of a random array with 0

```python
array = np.array(np.random.randint(0, 12+1, size=(1, 10)))
print(array)

# works for any array (since not specified, I replace all occurrences of the maximum value with 0)
zeroArray = np.where(array == np.max(array), 0, array)
print(zeroArray)
```

**3.** Construct the direct product of arrays (all combinations with each element). A two-dimensional array is provided as input.

```python
array = np.array(np.random.randint(0, 12+1, size=(2, 3)))
print(array)

arrCom = np.array(np.meshgrid(array[0], array[1])).T.reshape(-1, 2)
print(f"Combinations: {arrCom}")
```

**4.** Given two arrays, A (8x3) and B (2x2), find rows in A that contain elements from each row in B, regardless of the order of the elements in B.

```python
arrayA = np.array(np.random.randint(0, 12+1, size=(8, 3)))
print(f"Array A: {arrayA}")
arrayB = np.array(np.random.randint(0, 12+1, size=(2, 2)))
print(f"Array B: {arrayB}")

rowsOfAWithAnyFromEachRowOfB = np.array([row for row in arrayA if any(np.isin(row, arrayB))])
print(f"Rows of A with any from each row of B: {rowsOfAWithAnyFromEachRowOfB}")
```

**5.** Given a 10x3 matrix, find rows of unequal values ​​(for example, the row [2,2,3] remains, the row [3,3,3] is deleted)

```python
array = np.array(np.random.randint(0, 3, size=(10, 3)))
print(array)

withoutRowsWithEqualValues = np.array([row for row in array if np.unique(row).size > 1])
print(f"Array without rows with equal values: {withoutRowsWithEqualValues}")
```

**6.** Given a two-dimensional array, remove the rows that are repeated.

```python
array = np.array(np.random.randint(0, 3, size=(10, 2)))
print(array)

noRepetitions = np.unique(array, axis=0)
print(f"Array without repetitions: {noRepetitions}")
```

* __Task 1__: Calculate the product of nonzero elements on the diagonal of a rectangular matrix.  
 For example, for X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) answer is 3.

```python
#X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])
X = np.array(np.random.randint(0, 8, size=(3, 3)))
print(X)
#numpy
diagonal_product = np.prod(np.where(np.diagonal(X) != 0, np.diagonal(X), 1))
print(f"Diagonal product of X: {diagonal_product}")
#nonumpy
diagonal_product_nonp = 1
for i in range(len(X[0])):
    if X[i][i] != 0:
        diagonal_product_nonp *= X[i][i]
print(f"Diagonal product of X (nonumpy): {diagonal_product_nonp}")

```

* __Task 2__: Two vectors x and y are given. Check if they define the same multiset.  
  For example, for x = np.array([1, 2, 2, 4]), y = np.array([4, 2, 1, 2]) answer is True.

```python
X = np.array([1,2, 2, 4])
Y = np.array([4,2,1,2])


#numpy
npEqual = np.array_equiv(np.sort(X), np.sort(Y))
print(npEqual)   

#nonumpy
X = [1,2, 2, 4]
Y = [4,2,1,2]
nonNpEqual = (sorted(X) == sorted(Y))
print(nonNpEqual)
```

* __Task 3__: Find the maximum element in the vector x among the elements preceded by zero. 
 For example, for x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) answer is 5.

```python
#numpy
X = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])

zeroIndices = np.where(X == 0)[0]
zeroIndices = zeroIndices[zeroIndices < X.size-1]
maxAfterZero = np.max(X[zeroIndices+1])
print(maxAfterZero)

#nonumpy
X = [6, 2, 0, 3, 0, 0, 5, 7, 0]
maxAfterZeroNoNp = max([X[i+1] for i in range(len(X)-1) if X[i] == 0])
print(maxAfterZeroNoNp)
```

* __Task 4__: Implement run-length encoding. For some vector x, it is necessary to return a tuple of two vectors of the same length. The first one contains numbers and the second one contains how many times they need to be repeated.  
 For example, for x = np.array([2, 2, 2, 3, 3, 3, 5]) answer is (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
#numpy
X = np.array([2, 2, 2, 3, 3, 3, 5])

changeIndices = np.where(X[1:] != X[:-1])[0] + 1
indices = np.r_[0, changeIndices, X.size]
values = X[indices[:-1]]
lengths = np.diff(indices)
encoding = (values, lengths)

print(encoding)

#nonumpy
X = [2, 2, 2, 3, 3, 3, 5]
currVal = None
count = 0
values = []
lengths = []
for v in X:
    if v != currVal:
        values.append(v)
        if count != 0:
            lengths.append(count)
        count = 1
        currVal = v
    else:
        count+=1
lengths.append(count)
encoding = (values, lengths)
print(encoding)
```
