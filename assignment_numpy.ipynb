**1.** Given a random array, negate the sign of the elements whose values are between 3 and 8

```python

array = np.array(np.random.randint(0, 12+1, size=(1, 10)))
print(array)

# works for any array
nArray = np.where((array > 3) & (array < 8), np.negative(array), array)
print(nArray)
```

**2.** Replace the maximum element of a random array with 0

```python
import numpy as np
array = np.array(np.random.randint(0, 12+1, size=(1, 10)))
print(array)

# works for any array (since not specified, I replace all occurrences of the maximum value with 0)
zeroArray = np.where(array == np.max(array), 0, array)
print(zeroArray)
```
